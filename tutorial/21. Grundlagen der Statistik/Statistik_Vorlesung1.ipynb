## Messungen

* Experimentelle Messung: Wert ± Fehler (mit EInheiten!)
    * Kombiniere Daten, um das Ergebnis zu berechnen
    * Fehler berechnen
    * Mit Erwartungen vergleichen
    
Messung $g = 9.70 \pm 0.15\,\text{m s}^{-2}$

Erwartung $g = 9.81\,\text{m s}^{-2}$

### Fragen:

* Wie kommt man zum Ergebnis?
* Wie SCHÄTZT man die Fehler?
* Fehlerquellen?
    * Statistische (willkürliche Fluktuationen)
    * Systematische (Apparatur, Prozedur, ...)
        * Willkürlich (z.B. Interkalibration)
        * Verzerrung (z.B. Energieskala)
        * Vergessene/unbekannte Effekte
        
Wie genau kann man den Fehler schätzen?
* <10% ist schon gut $\to$ ≤2 signifikante Stellen für den Fehler!
* Ich möchte nie $g = 9.7034 ± 0.1545\,\text{m s}^{-2}$ sehen!

## Interpretation der Fehler (Unsicherheiten)

"Fehler" bedeutet nicht, dass man was falsch gemacht hat,
sondern die Unsicherheit auf der Messung

Messung $g = 9.70 \pm 0.15\,\text{m s}^{-2}$

* Ingenieur:
    * $9.55 ≤ g ≤ 9.85$
    * d.h. Fehler zeigt Toleranz oder erlaubten Bereich der Werte
* Physiker:
    * Experiment mehrmals wiederholen
    * $9.55 ≤ g ≤ 9.85$ 68% der Zeit
    * $9.40 ≤ g ≤ 10.00$ 95% der Zeit (wenn die Fehler Gaußsche sind)
    
*Warnung: Wenn man systematische Fehler evaluiert, Tendenz
ist häufig sie als Toleranz zu sehen und nicht als Gaußsche Fehler!!*

## Mittelwert

* Menge von nicht-gebinnte Daten (Messungen)
$\{ x_1, x_2, \ldots, x_N \}$
 
* Mittelwert:
$$\bar{x} = \frac{1}{N} \sum_{i=1}^{N} x_{i}$$

    * Wenn $f$ eine Funktion von $x$ ist, z.B. $f = ax$, $f = x^{2}$:
$$\bar{f} = \frac{1}{N} \sum_{i=1}^{N} f(x_{i})$$

* Für gebinnte Daten
    * $n_j$ Einträge im Bin
    * $x_j$ Bin-Mitte
    
<center><img src="figs/hist1.jpg" style="width: 400px"/></center>

$$\bar{x} = \frac{1}{N} \sum_{j=1}^{N_{\text{bin}}} n_{j} x_{j}\,;
\quad 
\bar{f} = \frac{1}{N} \sum_{j=1}^{N_{\text{bin}}} n_{j} f(x_{j})\,;
\quad
N = \sum_{j=1}^{N_{\text{bin}}} n_{j}
$$

## Breite charakterisieren

1. Idee?
$$\text{Breite} = \frac{1}{N} \sum_{i=1}^{N} (x_{i} - \overline{x})$$

2. Idee?
$$\text{Breite} = \frac{1}{N} \sum_{i=1}^{N} |x_{i} - \overline{x}|$$

### Varianz

$$V(x) = \frac{1}{N} \sum_{i=1}^{N} ( x_{i} - \overline{x} )^{2}
\qquad
V(f) = \frac{1}{N} \sum_{i=1}^{N} ( f(x_{i}) - \overline{f} )^{2}
$$

### Bei der Datenanalyse

$$
\begin{aligned}
V(x) & = \frac{1}{N} \sum_{i=1}^{N} ( x_{i} - \overline{x} )^{2}\\
  & = \frac{1}{N} \sum_{i=1}^{N} ( x_{i}^{2} - 2x_{i}\overline{x} +
  \overline{x}^{2} )\\
  & = \frac{1}{N} \sum_{i=1}^{N} x_{i}^{2} - 2 N \overline{x}^{2} +
  N \overline{x}^{2}\\
  & = \overline{x^{2}} - (\overline{x})^{2}
\end{aligned}
$$

da nur 1 Iteration über die Daten gebraucht wird.
Oft sinvoll einen grob geschätzten Mittelwert abzuziehen.

Iterative Formel

$$V_{N} = \frac{(N-1)}{N} V_{N-1} + \frac{(x - \overline{x}_{N-1})^{2}}{N}$$

### Standardabweichung

$$\sigma = \sqrt{V(x)} = \sqrt{\overline{x^{2}} - (\overline{x})^{2}}$$


## RMS und FWHM
 
* R.M.S. = Standardabweichung mit der obigen Definition
* FWHM (full width half maximum)
    * Nützlich für asymmetrische Verteilungen oder welche mit langen Schwänzen
    * Probleme mit Fluktuationen bei niedrigen Statistik 
    * Für eine Gaußverteilung: FWHM = 2.35 σ

## Kovarianz

Datensatz mit Paaren von Variablen: $\{ (x_1, y_1), (x_2, y_2), \ldots, (x_N, y_N) \}$
* Berechne $\overline{x}$, $\overline{y}$, $V(x)$, $V(y)$ wie vorher
* Kovarianz beschreibt ihre Abhängigkeit voneinander:
$$
\begin{aligned}
  \text{cov}(x,y) & =
  \frac{1}{N} \sum_{i} (x_{i} - \overline{x})(y_{i} - \overline{y}) \\
  & = \frac{1}{N} \sum_{i} x_{i}y_{i} -
  \frac{\sum x_{i}}{N} \frac{\sum y_{i}}{N} \\
  & = \overline{xy} - \overline{x} \cdot \overline{y}
\end{aligned}
$$
* Wenn $\overline{x y} = \overline{x} \cdot \overline{y}$ Variablen sind unabhängig
